```python
#boxplot give Interquartile range, median, outliers and whiskers. 
# The values outside the whiskers are considered as outliers.
# Wiskers are the lines that extend from the box to the minimum and maximum values of the dataset.
# The boxplot is a graphical representation of the distribution of a dataset. 
# It is a standardized way of displaying the distribution of data based on a five number 
# summary: minimum, first quartile, median, third quartile, and maximum.
#It is called Bi variate analysis because it is used to analyze the relationship between two variables.
gr=sns.boxplot(data=df,x='IncomeGroup',y='BirthRate') # it will show the boxplot of BirthRate for each IncomeGroup in a separate window with interactive features
```

```python
# it will show the scatter plot of BirthRate and InternetUsers with a 
# regression line in a separate window with interactive features
gr=sns.lmplot(x='BirthRate',y='InternetUsers',data=df)
```

```python
gr=sns.lmplot(x='BirthRate',y='InternetUsers',data=df,fit_reg=False) # it will show the scatter plot of BirthRate and InternetUsers without a regression line in a separate window with interactive features
```

```python
gr=sns.lmplot(x='InternetUsers',y='BirthRate',data=df,fit_reg=False,hue='IncomeGroup') # it will show the scatter plot of BirthRate and InternetUsers without a regression line and it will color the points based on the IncomeGroup in a separate window with interactive features
```

```python
gr=sns.lineplot(x='BirthRate',y='InternetUsers',data=df,)
```

```python
gr=sns.barplot(x='BirthRate',y='InternetUsers',data=df)
```

```python
gr=sns.catplot(x='BirthRate',y='InternetUsers',data=df,kind='bar') # it will show the bar plot of BirthRate and InternetUsers in a separate window with interactive features
```
