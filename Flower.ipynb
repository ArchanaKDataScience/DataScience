```python
import math
import time
import os

width, height = 80, 40

for t in range(100):
    os.system('cls' if os.name == 'nt' else 'clear')
    
    for y in range(height):
        for x in range(width):
            # Normalize coordinates
            nx = x - width / 2
            ny = y - height / 2
            
            r = math.sqrt(nx**2 + ny**2)
            angle = math.atan2(ny, nx)
            
            # Create wave pattern
            val = math.sin(r * 0.3 - t * 0.2 + angle)
            
            if val > 0.5:
                print("*", end="")
            elif val > 0:
                print(".", end="")
            else:
                print(" ", end="")
        print()
    
    time.sleep(0.05)
```
