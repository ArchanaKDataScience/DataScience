```python
#Stacked Bar Chart
A = [15., 30., 45., 22.]

B = [15., 25., 50., 20.]

z2 = range(4)

plt.bar(z2, A, color = 'b')
plt.bar(z2, B, color = 'r', bottom = A)

plt.show()
```

```python
#Adding a legend
x15 = np.arange(1, 5)

fig, ax = plt.subplots()

ax.plot(x15, x15*1.5)
ax.plot(x15, x15*3.0)
ax.plot(x15, x15/3.0)

ax.legend(['Normal','Fast','Slow']);
```

```python
x15 = np.arange(1, 5)

fig, ax = plt.subplots()

ax.plot(x15, x15*1.5, label='Normal')
ax.plot(x15, x15*3.0, label='Fast')
ax.plot(x15, x15/3.0, label='Slow')

ax.legend();
```

```python
#Control colours
# Colour abbreviation Colour name b blue c cyan g green k black m magenta r red w white y yellow
x16 = np.arange(1, 5)

plt.plot(x16, 'r')
plt.plot(x16+1, 'g')
plt.plot(x16+2, 'b')

plt.show()
```

```python
#Control line styles 
#Style abbreviation Style
# -   solid line
#-- dashed line
#-. dash-dot line
#: dotted line
x16 = np.arange(1, 5)

plt.plot(x16, '--', x16+1, '-.', x16+2, ':',x16+3, '-')

plt.show()
```
