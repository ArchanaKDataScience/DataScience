# superset , subset , disjoint

```python
a = {1,2,3,4,5,6,7,8,9} 
b = {3,4,5,6,7,8} 
c = {10,20,30,40} 
```

```python
a.issuperset(b)
```

```python
b.issubset(a)
```

```python
c.issub(a)
```

```python
print(a)
print(b)
print(c)
```

```python
c.isdisjoint(a) 
```

```python
a.isdisjoint(b)
```
